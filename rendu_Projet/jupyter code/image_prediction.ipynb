```python
# Manipulation de fichiers et de dossiers
import os

# Manipulation d'images
from PIL import Image

# Manipulation de matrices et d'images
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Machine learning
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Conv2DTranspose, BatchNormalization


# Base directory
base_dir = r"C:\Users\sofia\Desktop\4iabd\pa\data\image"

# Define parent directory
parent_dir = os.path.join(base_dir, 'maxj')

# Define output directory
output_dir = os.path.join(base_dir, 'prediction')
os.makedirs(output_dir, exist_ok=True)

# Define image directories
sub_dirs = ["NO2", "O3", "PM25", "PM10"]
image_dirs = [os.path.join(parent_dir, sub_dir) for sub_dir in sub_dirs]

# Define the coordinates for cropping
left = 144
top = 59
right = 510
bottom = 425

# Process each subdirectory
for sub_dir in image_dirs:

    # Process each file in the subdirectory
    for filename in os.listdir(sub_dir):
        # Check if the file is an image
        if filename.endswith(".jpg"):  
            # Get the file path
            file_path = os.path.join(sub_dir, filename)

            try:
                # Open the image
                img = Image.open(file_path)

                # Crop the image
                img_cropped = img.crop((left, top, right, bottom))

                # Save the cropped image
                output_path = os.path.join(sub_dir, filename)
                img_cropped.save(output_path)
                
            except Exception as e:
                print(f"Error processing file {file_path}: {str(e)}")

    print(f"Finished processing {sub_dir}\n")

def load_images_with_output(image_dir):
    # Load images
    image_files = os.listdir(image_dir)
    image_files = sorted(image_files)  # Make sure the images are in chronological order
    images = []

    # Load and resize images
    for image_file in image_files:
        if "_output.jpg" in image_file:
            image = cv2.imread(os.path.join(image_dir, image_file))
            image = cv2.resize(image, (256, 256))  # Resize to 256x256 pixels
            images.append(image)

    # Normalize pixel values to be between 0 and 1
    images = np.array(images) / 255.0

    return images, os.path.basename(image_dir)

for image_dir in image_dirs:
    # Load images with "_output.png" in their filenames
    images, gas_type = load_images_with_output(image_dir)

    # Prepare the data
    X = images[:-1]  # All images except the last one
    y = images[1:]  # All images except the first one

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)

    # Create the generator model
    generator = Sequential([
        Conv2D(64, kernel_size=(3, 3), strides=(2, 2), padding="same", activation="relu", input_shape=(256, 256, 3)),
        BatchNormalization(),
        Conv2D(128, kernel_size=(3, 3), strides=(2, 2), padding="same", activation="relu"),
        BatchNormalization(),
        Conv2DTranspose(64, kernel_size=(3, 3), strides=(2, 2), padding="same", activation="relu"),
        BatchNormalization(),
        Conv2DTranspose(3, kernel_size=(3, 3), strides=(2, 2), padding="same", activation="sigmoid"),
    ])

    # Compile the generator model
    generator.compile(optimizer="adam", loss="mean_squared_error")

    # Train the generator model
    generator.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=300)

    # Generate high-resolution images using the trained generator model
    predicted_images = generator.predict(X_test)

    # Save the predicted image to the output directory for each gas type
    output_file = os.path.join(output_dir, f"predicted_{gas_type}_output.png")
    plt.imshow(predicted_images[0])
    plt.savefig(output_file)

    # Display the first image from the test set and its predicted version
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(X_test[0])

    plt.subplot(1, 2, 2)
    plt.title("Predicted Image")
    plt.imshow(predicted_images[0])

    plt.show()


```
